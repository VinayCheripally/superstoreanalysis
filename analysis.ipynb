```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
df.info()
```

```python
#profit in each region
df.groupby("Region")["Profit"].sum().plot.pie(autopct='%1.1f%%')
```

```python
#profit in each category
df.groupby("Category")["Profit"].sum().plot.pie(autopct='%1.1f%%')
```

```python
#sales in each region
df.groupby("Region")["Sales"].sum().plot.pie(autopct='%1.1f%%')
```

```python
#sales in each category
df.groupby("Category")["Sales"].sum().plot.pie(autopct='%1.1f%%')
```

```python
#sales in each sub category
df.groupby("Sub-Category")["Sales"].sum().plot.bar()
```

```python
#profit in each sub category
df.groupby("Sub-Category")["Profit"].sum().plot.bar()
```

```python
#top ten states according to profit
df.groupby("State")["Profit"].sum().sort_values().nlargest(10).plot.bar()
```

```python
#top ten states according to sales
df.groupby("State")["Sales"].sum().sort_values().nlargest(10).plot.bar()
```

```python
#top ten cities according to sales
df.groupby("City")["Sales"].sum().sort_values().nlargest(10).plot.bar()
```

```python
#top ten cities according to profit
df.groupby("City")["Profit"].sum().sort_values().nlargest(10).plot.bar()
```

```python
#sales through each shiop mode
df.groupby("Ship Mode")["Sales"].sum().plot.pie(autopct='%1.1f%%')
```

```python
#profit through each ship mode
df.groupby("Ship Mode")["Profit"].sum().plot.pie(autopct='%1.1f%%')
```

```python
df.groupby("Segment")["Profit"].sum().plot.pie(autopct='%1.1f%%')
```

```python
df.groupby("Segment")["Sales"].sum().plot.pie(autopct='%1.1f%%')
```
